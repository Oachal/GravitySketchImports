Using Pandas to clean/manipulate the data
Using numpy and matplotlib for data visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Read in the comma delimted data file (Questions must be changed to Q# format)

```python
df = pd.read_csv ("C-LASSdataFormatFix.csv")
df.head(34)
```

The lines below will replace any NaN (Not any Number) values with -3. This is outside the range of our real data, but wont throw off the graphs...I think.

```python
df = df.fillna(-3)
df.head (30)
```

Apply the condition for the verification question (Answer 4 on Q31)

```python
df = df[df.Q31==1]
#df['Q31'] = df['Q31'].astype(int) Not needed, but good to know about
```

Next, seperate questions by factor:
PARRW: 3, 14, 25, 28, 30, 37
ESM: 23, 24, 29, 32
PSSE: 5, 21, 22, 25, 34, 40

```python
df_PARRW = df[['Q3','Q14','Q25','Q28','Q30','Q37']]
#df_tPARRW = df_PARRW.T #Why transpose these?!?
```

```python
df_ESM = df[['Q23','Q24','Q29','Q32']]
#df_tESM = df_ESM.T #It looks like these are unnecesary
```

```python
df_PSSE = df[['Q5','Q21','Q22','Q25','Q34','Q40']]
#df_tPSSE = df_PSSE.T
```

Experts Agree: 2,3,11,14,15,16,19,24,25,26,28,30,(31*),34,36,37,38,39,42
Experts Disagree:1,5,6,8,10,12,13,17,18,20,21,22,23,27,29,32,35,40
No Consensus: 4,7,9,33,41

Next we make plots of the grouped questions...Notched boxes represent confidence intervals/Quartiles of reponses.

```python
labels = ['Q3','Q14','Q25','Q28','Q30','Q37']
bp=plt.boxplot(df_PARRW,
            labels=labels,
            #notch=True,
            #bootstrap=None,
           patch_artist=True)

colors = ['lightgreen', 'lightgreen','lightgreen', 'lightgreen', 'lightgreen','lightgreen']

for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

plt.ylabel('Student Response')
plt.show
```

```python
labels = ['Q23','Q24','Q29','Q32']
bp=plt.boxplot(df_ESM,
            labels=labels,
            #notch=True,
            #bootstrap=None,
           patch_artist=True)

colors = ['red', 'lightgreen','red', 'red']

for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

plt.ylabel('Student Response')
plt.show
```

```python
labels = ['Q5','Q21','Q22','Q25','Q34','Q40']
bp=plt.boxplot(df_PSSE,
            labels=labels,
            #notch=True,
            #bootstrap=None,
           patch_artist=True)

colors = ['red', 'red','red', 'lightgreen', 'lightgreen','red']

for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

plt.ylabel('Student Response')
plt.show
```
